```python
from langdetect import detect
```

```python
def detect_lang(text):
    try: 
        return detect(text)
    except Exception:
        return 'exp'
```

```python
tweets['lang'] = tweets['tweet'].map(lambda t: detect_lang(t))
```

```python
tweets.head()
```

```python
tweets = tweets[tweets.lang == 'en']
```

```python
#tweets.to_csv('Uber_Dataset.csv')
```

```python
import re
def remove_Names(tweet):  
    return re.sub('@[^\s]+','',tweet)
```

```python
tweets['tweet']=tweets['tweet'].apply(lambda tweet:remove_Names(tweet))
```

```python
tweets.head()
```

```python
def remove_Hashtags(tweet):  
    return re.sub('#[^\s]+','',tweet)
```

```python
tweets['tweet']=tweets['tweet'].apply(lambda tweet:remove_Hashtags(tweet))
```

```python
tweets.head()
```

```python
tweets.drop("lang",axis=1,inplace=True)
```

```python
import string
def remove_punct(text):
    text  = "".join([char for char in text if char not in string.punctuation])
    text = re.sub('[0-9]+', '', text)
    return text

```

```python
tweets['tweet'] = tweets['tweet'].apply(lambda x: remove_punct(x))
tweets.head(10)
```

```python
from textblob import TextBlob
def getSubjectivity(text):
    return TextBlob(text).sentiment.subjectivity

# Create a function to get the polarity
def getPolarity(text):
    return  TextBlob(text).sentiment.polarity
```

```python
# Create two new columns 'Subjectivity' & 'Polarity'
tweets['Subjectivity'] = tweets['tweet'].apply(getSubjectivity)
tweets['Polarity'] = tweets['tweet'].apply(getPolarity)

# Show the new dataframe with columns 'Subjectivity' & 'Polarity'
tweets.head()
```

```python
from wordcloud import WordCloud, STOPWORDS 
import matplotlib.pyplot as plt 
import pandas as pd 
```

```python
allWords = ' '.join([twts for twts in tweets['tweet']])
wordCloud = WordCloud(width=500, height=300, random_state=21, max_font_size=110).generate(allWords)
plt.imshow(wordCloud, interpolation="bilinear")
plt.axis('off')
plt.show()
```

```python
# Create a function to compute negative (-1), neutral (0) and positive (+1) analysis
def getAnalysis(score):
    if score < 0:
        return 'Negative'
    elif score == 0:
        return 'Neutral'
    else:
        return 'Positive'
tweets['Analysis'] = tweets['Polarity'].apply(getAnalysis)
tweets.head()

```

```python
tweets.head(20)
```

```python
# Printing positive tweets 
print('Printing positive tweets:\n')
j=1
sortedDF = tweets.sort_values(by=['Polarity'],ascending=False) #Sort the tweets
j=1
sortedDF.reset_index(drop=True, inplace=True)
for i in range(0, sortedDF.shape[0] ):
    if( sortedDF['Analysis'][i] == 'Positive'):
        print(str(j) + ') '+ sortedDF['tweet'][i])
        print()
        j= j+1
```

```python
# Printing negative tweets  
print('Printing negative tweets:\n')
j=1
for i in range(0, sortedDF.shape[0] ):
      if( sortedDF['Analysis'][i] == 'Negative'):
        print(str(j) + ') '+sortedDF['tweet'][i])
        print()
        j=j+1
```

```python
# Plotting 
plt.figure(figsize=(8,6)) 
for i in range(0, sortedDF.shape[0]):
      plt.scatter(sortedDF["Polarity"][i], sortedDF["Subjectivity"][i], color='Blue') 
# plt.scatter(x,y,color)   
plt.title('Sentiment Analysis') 
plt.xlabel('Polarity') 
plt.ylabel('Subjectivity') 
plt.show()
```
